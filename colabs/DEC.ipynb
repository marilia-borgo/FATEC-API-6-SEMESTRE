# Indicadores-continuidade-coletivos-2020-2029

```python
import requests
import pandas as pd

# Filtrei em ambas a URL o conjunto GUACUI como amostra, seu IdeConjUndConsumidoras = 15394
URL_DEC_REALIZADO = 'https://dadosabertos.aneel.gov.br/api/3/action/datastore_search_sql?sql=SELECT%20*%20FROM%20%224493985c-baea-429c-9df5-3030422c71d7%22%20WHERE%20%22IdeConjUndConsumidoras%22%20IN%20(%2715394%27)'
URL_DEC_LIMITE = 'https://dadosabertos.aneel.gov.br/api/3/action/datastore_search_sql?sql=SELECT%20*%20FROM%20%22fd69e1dd-fd66-4269-b60c-cc0b7eb221b4%22%20WHERE%20%22IdeConjUndConsumidoras%22%20IN%20(%2715394%27)'


# Carrega os dados no dataframe
response_dec_realizado = requests.get(URL_DEC_REALIZADO, timeout=30)
records_dec_realizado = response_dec_realizado.json()['result']['records']
df_dec_realizado = pd.DataFrame(records_dec_realizado)

print(df_dec_realizado.shape)
df_dec_realizado
```

```python
response_dec_limite = requests.get(URL_DEC_LIMITE, timeout=30)
records_dec_limite = response_dec_limite.json()['result']['records']
df_dec_limite = pd.DataFrame(records_dec_limite)

print(df_dec_limite.shape)
df_dec_limite
```

# Tratamento de Dados

```python
# Converte a coluna para número (trocando vírgula por ponto se necessário)
df_dec_realizado['VlrIndiceEnviado'] = pd.to_numeric(
    df_dec_realizado['VlrIndiceEnviado']
    .astype(str)
    .str.replace(',', '.', regex=False),
    errors='coerce',
)

df_dec_limite['VlrLimite'] = pd.to_numeric(
    df_dec_limite['VlrLimite'].astype(str).str.replace(',', '.', regex=False),
    errors='coerce',
)
```

```python
df_dec_realizado_filtro1 = df_dec_realizado[
    df_dec_realizado['SigIndicador'] == 'DEC'
]

# Agrupamento DEC REALIZADO
df_dec_realizado_agrupado = (
    df_dec_realizado_filtro1
    .groupby(
        ['IdeConjUndConsumidoras', 'DscConjUndConsumidoras'], as_index=False
    )['VlrIndiceEnviado']
    .sum()
    .sort_values('VlrIndiceEnviado', ascending=False)
)
df_dec_realizado_agrupado
```

```python
df_dec_limite_filtro1 = df_dec_limite[df_dec_limite['SigIndicador'] == 'DEC']

# Agrupamento DEC LIMITE
df_dec_limite_agrupado = (
    df_dec_limite_filtro1
    .groupby(
        ['IdeConjUndConsumidoras', 'DscConjUndConsumidoras'], as_index=False
    )['VlrLimite']
    .sum()
    .sort_values('VlrLimite', ascending=False)
)

df_dec_limite_agrupado
```

# Resultado

```python
df_resultado = pd.merge(
    df_dec_realizado_agrupado,
    df_dec_limite_agrupado,
    on='IdeConjUndConsumidoras',
)
df_resultado

# Remover coluna
df_resultado = df_resultado.drop(columns=['DscConjUndConsumidoras_y'])

df_resultado_renomeado = df_resultado.rename(
    columns={
        'DscConjUndConsumidoras_x': 'Conjunto Elétrico',
        'VlrIndiceEnviado': 'DEC_Realizado',
        'VlrLimite': 'DEC_Limite',
    }
)

df_resultado_renomeado
```
